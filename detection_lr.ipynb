```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
model = LogisticRegression(penalty="l2",C=1.0,max_iter=1000,random_state=42)
```

```python
df = pd.read_csv('/content/drive/MyDrive/ML-Phishing Detection Project/data/url_features.csv')
data = df.drop(columns=['url','protocol','domain','path','query','fragment'])
```

```python
x = data.drop(["phishing"],axis=1)
x
y = data[["phishing"]]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y ,random_state=42,stratify=y)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
y_train
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
accuracy = accuracy_score(y_test,y_pred)*100

confusion_mat = confusion_matrix(y_test,y_pred)
print("Accuracy is",accuracy)
print("Confusion Matrix")
print(confusion_mat)
```
